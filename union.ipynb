```python
import cv2 
import numpy as np 

# 좌우 이미지 불러오기 
for a in range(20):
    aa=a
    aas=str(aa)
    road="./new_crop/0628/results_video"+aas
    leftimg = cv2.imread('./new_crop/0628/'+aas+'.png') 

    rightimg = cv2.imread(road+'/0_re.png') 
    # 이미지 좌우 합치기
    add_img = np.hstack((leftimg, rightimg))
    for i in range(5):
        num=(i+1)*50
        nums=str(num)
        leftimg=add_img
        rightimg=cv2.imread(road+'/'+nums+'_re.png')
        add_img = np.hstack((leftimg, rightimg))
    cv2.imwrite("./new_crop/0628/new"+aas+"_1.png", add_img);
```
